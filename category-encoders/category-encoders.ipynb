```python
import numpy as np
import pandas as pd
from category_encoders.ordinal import OrdinalEncoder
from sklearn.datasets import load_boston
```

```python
b = load_boston()
X = pd.DataFrame(b.data, columns=b.feature_names)
X
```

```python
X["RAD"].unique()
```

```python
X_train = X.iloc[:400, :]
X_valid = X.iloc[400:, :]
```

**category-encoders は pd.DataFrame のまま OrdinalEncoder に渡せる**

```python
enc = OrdinalEncoder()
train_rad_encoded = enc.fit_transform(X_train["RAD"])
train_rad_encoded
```

```python
valid_rad_encoded = enc.transform(X_valid["RAD"])
valid_rad_encoded
```

```python
X_valid["rad_encoded"] = valid_rad_encoded
X_valid
```

**sklearn の LabelEncoder のように np.array に変換してから渡しても pd.DataFrame で return してくる（当然 index はリセットされてる）ので注意**

```python
valid_rad_encoded_ = enc.transform(X_valid["RAD"].values)
valid_rad_encoded_
```

**index が元データフレームとズレているので列追加しても全て NaN になってしまう**

**当然レコード数は同じなのでエラーを吐かないので確認しないと気づかないので注意**

```python
X_valid["rad_encoded_"] = valid_rad_encoded_
X_valid
```
